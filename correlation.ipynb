```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("istherecorrelation.csv", delimiter = ";")
df["WO"] = df["WO [x1000]"].str.replace(",", ".").astype("float")
df["Beer consumption (hl)"] = df["NL Beer consumption [x1000 hectoliter]"] * 1000
```

```python
plt.figure(figsize = (10, 6))
plt.scatter(df["WO"], df["Beer consumption (hl)"])
correlaltion = df["WO"].corr(df["Beer consumption (hl)"])
plt.text(df["WO"].min(), df["Beer consumption (hl)"].max(), f"Correlation: {correlaltion:.2f}", fontsize=12, verticalalignment='top')
z = np.polyfit(df["WO"], df["Beer consumption (hl)"], 1)
p = np.poly1d(z)
plt.plot(df["WO"], p(df["WO"]), color='blue', linestyle='--', label='fitted trendline ')
plt.legend()
plt.xlabel("WO Students [x1000]")
plt.ylabel("Beer consumption (hl)")
plt.title("Correlation between WO students and beer consumption")
plt.grid()
plt.savefig("correlation_graph", dpi = 300)
plt.show()
```
